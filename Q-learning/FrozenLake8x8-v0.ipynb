```python
import numpy as np
import gym
import random
```

```python
env = gym.make("FrozenLake8x8-v0")
```

```python
action_size = env.nA
state_size = env.nS
```

```python
print("Total Action {}".format(action_size))
print("Total State {}".format(state_size))
```

```python
qtable = np.zeros((state_size, action_size))
```

```python
# hyper parameters
total_episodes = 200000
max_steps = 256
gamma = 0.95
alpha = 0.8 # learning rate

#exploration parameters
epsilon = 1
min_epsilon = 0.01
max_epsilon = 1.0
decay_rate = 0.00001
```

```python
# Training Stage
rewards = []
for episode in range(total_episodes * 2):
    
    state = env.reset()
    total_reward = 0
    done = False
    
    for step in range(max_steps):
#         trade_off  = random.uniform(0, 1)
        
#         if trade_off > epsilon:
#             action = np.argmax(qtable[state, :])
#         else:
#             action = env.action_space.sample()
        action  = env.action_space.sample()
        
        new_state, reward, done, info = env.step(action)
        #print("{} # step {} : {}".format(episode, step, reward))
        qtable[state, action] = qtable[state, action] + alpha * (reward + gamma * np.max(qtable[new_state, :]) - qtable[state, action])
        
        total_reward += reward
        
        state = new_state
        
        if done == True:
            break
            
    epsilon = min_epsilon + (max_epsilon - min_epsilon) * np.exp(-decay_rate * episode)
    rewards.append(total_reward)
    #rint("{} # epsilon : {}".format(episode, epsilon))
    #rint(" ", trade_off > epsilon)

print("Score Over Time {}".format(sum(rewards)/ total_episodes))
print("Q table :")
print(qtable)
```

```python
for episode in range(total_episodes):
    
    state = env.reset()
    total_reward = 0
    done = False
    
    for step in range(max_steps * 2):
        
        action = np.argmax(qtable[state, :])
        
        new_state, reward, done, info = env.step(action)
        #print("{} # step {} : {}".format(episode, step, reward))
        qtable[state, action] = qtable[state, action] + alpha * (reward + gamma * np.max(qtable[new_state, :]) - qtable[state, action])
        
        total_reward += reward
        
        state = new_state
        
        if done == True:
            break
            
    rewards.append(total_reward)

print("Score Over Time {}".format(sum(rewards)/ total_episodes))
print("Q table :")
print(qtable)
```

```python
# Testing Stage

state = env.reset()
done = False

for step in range(max_steps):

    action = np.argmax(qtable[state, :])

    new_state, reward, done, info = env.step(action)
    
    
    if done == True:
        env.render()
        print("It has taken {} steps.".format(step + 1))
        break
    state = new_state
    
env.close()
```
